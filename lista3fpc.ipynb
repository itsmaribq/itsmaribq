```python
#EXERCICIO 5 (OPCIONAL)
import numpy as np

# Define uma função chamada jacobi
def jacobi(A, b, tol=1e-10):#A:matriz dos coeficientes do sist
#b:Vetor do lado direito da equação
#tolerancia para a parara 1.10^-10

    n = len(b)#Calcula o número de incógnitas no sistema de equações

    #criando um vetor x de soluções
    x = np.zeros_like(b, dtype=float)#np.zeros cria um vetor do mesmo tamanho de b que é preenchido com zeros e depois vai armazenar os valores das iterações

    # Iteração Jacobi
    for _ in range(100):#crio um for que vai iterar até no max 100 vezes
        # Inicializa uma nova solução x_new
        x_new = np.zeros_like(x, dtype=float)#criando um novo vetor para armazenar as estimativas atualizadas, inicialmente é um vetor de zeros

        for i in range(n):#criei um loop que percorre cada incógnita i do sistema

            s = sum(A[i][j] * x[j] for j in range(n) if j != i)#Calcula a soma s dos elementos que não estão na diagonal da matriz A pois esses estão sendo tratados em outra parte

            # Atualiza x_new com base na fórmula do método de Jacobi
            x_new[i] = (b[i] - s) / A[i][i]

        #criterio de parada
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:#Calcula a diferença entre o vetor x_new (novo) e x (anterior) usando a norma infinita a qual vai verificar a diferença absoluta ent're os elementos
            break

        # Atualiza x para a próxima iteração
        x = x_new

    return x

```

```python
#MINI PROJETO pt1

import numpy as np


def dens_const(x):##Define uma função de densidade constante ao longo da corda
    return 1  #nesse caso, para qualquer valor de x, a densidade é 1 kg/m


def dens_var(x):##Definindo a densidade variável que depende da posição de x
    return 0.5 * (1 + np.exp(-100 * (x - 0.5)**2))  # kg/m
#A densidade varia suavemente ao longo da corda, aumentando no meio e diminuindo nas extremidades

# Função para calcular a massa total da corda
def calculo_massa(N, densidade_func):#Essa função divide a corda em N intervalos igualmente espaçados
    dx = 1 / N  # Comprimento de cada intervalo
    pontos = np.linspace(0, 1, N)  # crio N pontos igualmente espaçado entre o,1/é nescessário que sejam igualmente espaçados pois caso contrario o deltax não seria costante e o calculo daria errado
    massa_total = sum(densidade_func(x) * dx for x in pontos)#a massa total é calculada fazendo a soma das massas de cada intervalo /Esta função avalia qual é a densidade no ponto específico x
   #dx representa o compr de cada segm, for x in pon itera sobre todos os pontos igualmente espaçados
    return massa_total

# Valores de N para testar
lista_N = [10, 20, 40, 80]#número de divisões da corda

# Calculando e exibindo a massa para cada valor de N e para cada função de densidade
for N in lista_N:
    massa_constante = calculo_massa(N, dens_const)
    massa_variavel = calculo_massa(N, dens_var)
    print(f"Valor de N = {N}")
    print(f"Massa com densidade constante: {massa_constante:.4f} kg")
    print(f"Massa com densidade variável: {massa_variavel:.4f} kg\n")
```

```python
#MINI PROJETO PT2 DEFINITIVO
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Função para densidade constante
def densidade_constante(x):  # Definindo a densidade constante
    return 1  # Retorna uma densidade constante de 1 kg/m (independente de x)

# Função para densidade variável
def densidade_variavel(x):  # Definindo a densidade variável
    # A densidade depende de x
    return 0.5 * (1 + np.exp(-100 * (x - 0.5) ** 2))  # Também em kg/m
    #a função de densidade inclui um termo gaussiano centrado em 0.5, portanto tem seu maximo em 0.5

# Função para calcular as massas de cada segmento da corda
def calc_massas(L, N, densidade):

    l_0 = L / N  # Comprimento de cada segmento da corda

    massas = []  # Lista para armazenar as massas de cada segmento

    for i in range(N):#um loop for para iterar sobre os índices de cada segmento da corda.
        a = (i + 0.5) * l_0  # Limite inferior do segmento i um pouco antes do centro do segmento.
        b = (i + 1.5) * l_0  # Limite superior do segmento i um pouco depois do centro do segmento.
        # Calcula a massa do segmento i usando a integral da densidade

        massa_i, _ = quad(densidade, a, b)  #Calcula a integral da função de densidade entre os limites a e b
        massas.append(massa_i)  # Adiciona a massa calculada à lista de massas
    return massas  # Retorna a lista de massas

# Função para calcular a posição de equilíbrio das massas usando o método iterativo de Jacobi
def jacobi(L, N, T, g, densidade, f):
    l_0 = L / N  # Comprimento de cada segmento da corda
    y = [0] * (N + 1)  # Lista de deslocamentos para cada ponto ao longo da corda, inicialmente em zero/uma corda dividida em N segmentos terá N + 1 pontos
    y_new = [0] * (N + 1)  # Lista para armazenar os novos valores de deslocamento após cada iteração

    # Calcula a lista de massas de cada segmento usando a função calc_massas
    massas = calc_massas(L, N, densidade)

    # Configurações iniciais: fixar as posições nas extremidades da corda/independentemente das forças que atuam no resto da corda, as extremidades não se movem
    y[0] = 0  # A extremidade esquerda é fixa, y[0] = 0
    y[N] = 0  # A extremidade direita é fixa, y[N] = 0

    # Critério de parada: erro menor que epsilon para considerar a convergência
    epsilon = 1e-6  # Precisão desejada para o critério de parada
    max_iter = 1500  # Número máximo de iterações para evitar loop infinito caso não converja

    # Início do loop de iterações de Jacobi
    for k in range(max_iter):
        max_diff = 0  # Variável para armazenar a maior diferença entre iterações (para critério de parada)

        # Iteração sobre cada ponto interno da corda (excluindo as extremidades fixas)
        for i in range(1, N):
            # Atualiza o deslocamento do ponto i usando a fórmula do método de Jacobi
            y_new[i] = ((f - massas[i] * g) * l_0 + T * (y[i - 1] + y[i + 1])) / (2 * T)

            # Calcula a diferença absoluta entre o novo e o antigo valor de y para critério de parada
            max_diff = max(max_diff, abs(y_new[i] - y[i])) #?


        # Atualiza a lista y com os novos valores calculados
        y = y_new[:]

        # Verifica o critério de convergência (se a maior diferença é menor que epsilon)
        if max_diff < epsilon:
            break  # Se o critério for atendido, encerra o loop de iterações

    return y  # Retorna a lista final de deslocamentos calculados

# Função para plotar gráficos da forma de equilíbrio da corda para diferentes densidades e valores de N
def Plotagem(N_values, densidades):
    L = 1.0  # Comprimento total da corda em metros
    T = 2.0  # Tensão aplicada à corda em Newtons
    g = 9.8  # Aceleração da gravidade em m/s²
    f = 0    # Força adicional externa (considerada zero neste caso)

    # Cria uma figura de tamanho especificado para o gráfico
    plt.figure(figsize=(10, 6))

    # Itera sobre cada densidade especificada (constante e variável)
    for densidade in densidades:
        # Define o texto da legenda com base no tipo de densidade
        if densidade == densidade_constante:
            label_base = 'Densidade Constante'
        elif densidade == densidade_variavel:
            label_base = 'Densidade Variável'

        # Itera sobre cada valor de N para criar gráficos comparativos
        for N in N_values:
            # Calcula a posição de equilíbrio usando o método de Jacobi
            y_equilibrio = jacobi(L, N, T, g, densidade, f)

            # Cria uma lista de posições ao longo do comprimento da corda (eixo x)
            x_posicao = np.linspace(0, L, N + 1)

            # Adiciona uma curva ao gráfico para o valor atual de N
            plt.plot(x_posicao, y_equilibrio, label=f'{label_base}, N = {N}')

    # Configurações do gráfico: título, rótulos dos eixos, legenda e grade
    plt.title(f'Forma de Equilíbrio da Corda')
    plt.xlabel('Comprimento da Corda (m)')
    plt.ylabel('Deslocamento de Equilíbrio (m)')
    plt.legend()
    plt.grid(True)
    plt.show()  # Exibe o gráfico gerado

# Definindo os valores de N (número de segmentos da corda) que serão usados nos gráficos
N_values = [10, 20, 40, 80]  # Diferentes valores de N para testar a convergência
densidades = [densidade_constante, densidade_variavel]  # Lista das funções de densidade a serem testadas

# Chama a função de plotagem para gerar os gráficos com os parâmetros especificados
Plotagem(N_values, densidades)
```

```python
#PARTE 3 (GAUSS-SEIDEL)

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Função de densidade constante
def densidade_constante(x):
    return 1  # kg/m

# Função para calcular as massas das partículas
def calc_massas(L, N, densidade):
    l_0 = L / N #Definindo l_0
    Lista_massas = [] #Lista vazia para colocar as massas calculadas na integral

    for i in range(N): #Rodando um for para o valor de N  ?
        a = (i + 0.5) * l_0  # Limite inferior da integral
        b = (i + 1.5) * l_0  # Limite superior da integral
        m_i, _ = quad(densidade, a, b) #Igualando as massas
        Lista_massas.append(m_i) #Colocando as massas na lista?
    return Lista_massas #Retornando os valores da Lista

# Método de Gauss-Seidel com força variável f_39 aplicada na partícula 39
def gauss_seidel_forca_variavel(L, N, T, g, densidade, f_39):#atualiza iterativamente os deslocamentos de cada partícula, considerando a tensão da corda e a gravidade
#calcula a massa para cada partícula ao longo da corda, integrando a densidade entre os limites especificados

    l_0 = L / N #Calculo de l_0
    y = [0] * (N + 1) ## Lista para armazenar deslocamentos das partículas (inicialmente zero)

    # Calcula as massas das partículas internas
    massas = calc_massas(L, N, densidade)

    # Condições de contorno fixas
    y[0] = 0  # Posição fixa na extremidade esquerda
    y[N] = 0  # Posição fixa na extremidade direita

    # Iteração do método de Gauss-Seidel
    for k in range(1000): #realizando até 1000 iterações para garantir que o sistema atinja um estado próximo ao equilíbrio
        for i in range(1, N): #loop interno, que itera sobre cada partícula interna da corda menos as extremidades/atualiza o deslocamento das particulas internas

            # Aplica a força variável em i = 39
            if i == 39: # Se a posição, i for a 39, colocamos uma força variável nela
                f = f_39 #Força na posição 39
            else:
                f = 0 #Caso não seja, a força será igual a zero
            # Atualiza y[i] usando o valor atualizado de y[i-1] e o valor não atualizado de y[i+1]
            y[i] = (f - massas[i] * g) * l_0 / (2 * T) + 0.5 * (y[i - 1] + y[i + 1]) #Método de gauss, explicação está no código acima

    return y[39]  # Retorna o deslocamento do ponto i = 39

# Parâmetros para o cálculo
L = 1.0  # Comprimento da corda em metros
T = 2.0  # Tensão da corda em Newtons
g = 9.8  # Aceleração da gravidade em m/s²
N = 80   # Número de segmentos

# Intervalo de valores para a força f_39
f_values = np.linspace(0, 10, 50)  # Testar valores de força de 0 a 10 N #A força é variável e vai de 0 a 10, com 50 "divisões"
deslocamentos = [] #Lista para colocarmos a posição deslocamento (y_39)

# Calcular o deslocamento para cada valor de f_39
for f_39 in f_values: #Calculamos o deslocamento de acordo com a força variável
    deslocamento = gauss_seidel_forca_variavel(L, N, T, g, densidade_constante, f_39) #Igualando deslocamento a função de achar o deslocamento
    deslocamentos.append(deslocamento) #Colocando o deslocamento na lista de deslocamento

# Plotar o deslocamento em função da força f_39
plt.figure(figsize=(10, 6))
plt.plot(f_values, deslocamentos, label='Deslocamento de y_39')
plt.xlabel('Força f_39 (N)')
plt.ylabel('Deslocamento de y_39 (m)')
plt.title('Deslocamento do Ponto i=39 em Função da Força Vertical f_39')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#PARTE 4
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad


def densidade_constante(x): #Definindo a função constante
    return 1  # Para qualquer valor de x, a densidade será sempre 1 kg/m
    #Essa função é usada para calcular as massas dos segmentos da corda

# Função para calcular as massas das partículas
def calc_massas(L, N, densidade): #Função que calcula a massa, seus argumentos são L, N numero de segmentos da corda e a densidade
    l_0 = L / N #Definindo l_0
    Lista_massas = [] #Lista vazia para colocar as massas

    for i in range(N): #Roda um for para cada valor de N
        a = (i + 0.5) * l_0  # Limite inferior da integral
        b = (i + 1.5) * l_0  # Limite superior da integral
        m_i, _ = quad(densidade, a, b) #Calcula a massa por meio da integral da função densidade com os limites em a e b
        Lista_massas.append(m_i) #Adicionando as massas na lista
    return Lista_massas #Retorna a lista de massas calculadas

# Método de Gauss-Seidel com duas forças aplicadas em diferentes pontos
def gauss_seidel_duas_forcas(L, N, T, g, densidade, f_17, f_30): #Definindo a função para o calculo das posições de equilibrio com a força em i = 17 e i = 30
    l_0 = L / N #Definindo l_0
    y = [0] * (N + 1) #Lista com N zeros/Como a corda é dividida em N segmentos, você tem N + 1 pontos ou partículas

    # Calcula as massas das partículas internas
    massas = calc_massas(L, N, densidade) #Calcula as massas das particulas

    # Condições de contorno fixas
    y[0] = 0  # Posição fixa na extremidade esquerda
    y[N] = 0  # Posição fixa na extremidade direita

    # Iteração do método de Gauss-Seidel
    for k in range(1000): #Iterando 1000 vezes para convergência
        for i in range(1, N):  #For que vai de 1 até o valor N
            # Aplica as forças nas posições i = 17 e i = 30 Positiva, para cima
            if i == 17: #Se i for 17, a força na fórmula séria f_17
                f = f_17  # Força para cima em i = 17
            elif i == 30: #Se i for 30, a força na fórmula será -f_30 Por que ela é contrária, negativa, para baixo
                f = -f_30  # Força para baixo em i = 30
            else:
                f = 0 #Se não for nenhuma dessas a força será zero
            # Atualiza y[i] usando o valor atualizado de y[i-1] e o valor não atualizado de y[i+1]
            y[i] = (f - massas[i] * g) * l_0 / (2 * T) + 0.5 * (y[i - 1] + y[i + 1]) #Cálculo de y[i] por meio do método de gauss, explicação está acima no exercio anterior

    return y  # Retorna a forma de equilíbrio ao longo da corda

# Parâmetros para o cálculo
L = 1.0  # Comprimento da corda em metros
T = 2.0  # Tensão da corda em Newtons
g = 9.8  # Aceleração da gravidade em m/s²
N = 80   # Número de segmentos

# Valores das forças aplicadas
f_17 = 10  # Força para cima em i = 17
f_30 = 10 # Força para baixo em i = 30

# Calcular a forma de equilíbrio com as forças aplicadas
y_equilibrio = gauss_seidel_duas_forcas(L, N, T, g, densidade_constante, f_17, f_30) #Calculando as posições de equilibro y

# Posições ao longo da corda
x_positions = np.linspace(0, L, N + 1) #Posição x, vai de 0 até L com N+1 pontos

# Plotar a forma de equilíbrio da corda
plt.figure(figsize=(10, 6))
plt.plot(x_positions, y_equilibrio, label='Forma de Equilíbrio com Forças em i=17 e i=30')
plt.xlabel('Comprimento da Corda (m)')
plt.ylabel('Deslocamento de Equilíbrio (m)')
plt.title('Forma de Equilíbrio da Corda com Forças em i=17 e i=30')
plt.legend()
plt.grid(True)
plt.show()
```
