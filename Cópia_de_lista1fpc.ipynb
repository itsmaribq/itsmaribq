```python
import numpy as np
import random
import time
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
#EXERCICIO 1 PT1
#é nescessário definir uma função para calcular o vetor reultante
def calcular_vetor_resultante(n):
    #gerar os vetores aleatorios usando numpy
    a = np.random.rand(n)
    b = np.random.rand(n)

    #definindo alpha e beta como números aleatórios
    alpha = np.random.rand()
    beta = np.random.rand()

    #para calcular o vetor c
    c = alpha*a + beta*b
    #devolvendo o vetor c que foi calculado pela formúla
    return c

#chamando a função
n = 5
c = calcular_vetor_resultante(n)
print("Vetor resultante c:", c)
```

```python
#EXERCIO 1 PT2

# Função para calcular a potência de uma matriz e medir o tempo gasto
def calcular_tempo_potencia(n, m):
    A = np.random.rand(n, n)  # função para gerar matriz aleatória de dimensão n x n
    start_time = time.time()  # Iniciar cronômetro
    A_m = np.linalg.matrix_power(A, m)  # Calcula A^m
    end_time = time.time()  # a subtração dos dois representa o tempo total gasto em A ^m
    return end_time - start_time  # Retorna o tempo gasto

# Dimensões e valores de m a serem testados
dimensoes = [10**2, 10**3, 10**4]
valores_m = [2, 3, 4]  # Valores de m (potência)

# Listas para armazenar os resultados
tempos = {m: [] for m in valores_m}

# Medir o tempo para cada dimensão e valor de m
for n in dimensoes:
    for m in valores_m:
        tempo_gasto = calcular_tempo_potencia(n, m)
        tempos[m].append(tempo_gasto)
        print(f"Tempo para n={n}, m={m}: {tempo_gasto:.4f} segundos")

# Plotar os resultados
plt.figure(figsize=(10, 6))

# Plot com escala linear
for m in valores_m:
    plt.plot(dimensoes, tempos[m], label=f'm={m}', marker='o')

plt.xlabel('Dimensão n')
plt.ylabel('Tempo (s)')
plt.title('Tempo de cálculo de A^m para diferentes valores de n e m')
plt.legend()
plt.grid(True)
plt.show()

# Plot com escala log-log
plt.figure(figsize=(10, 6))
for m in valores_m:
    plt.plot(dimensoes, tempos[m], label=f'm={m}', marker='o')

plt.xscale('log')
plt.yscale('log')
plt.xlabel('Dimensão n (log)')
plt.ylabel('Tempo (s) (log)')
plt.title('Tempo de cálculo de A^m em escala log-log')
plt.legend()
plt.grid(True)
plt.show()
```

```python
 #EXERCICO 2
# Função para multiplicação de matrizes quadradas e medição do tempo
def multiplica_matriz(dimensao): #define a função que recebe o argumento dimensions que consiste em uma lista com diferentes dimensões para as matrizes quadradas que vão ser multiplicadas  cada valor da lsita representa um tamanho n de uma matriz nxn
    times = [] #lista vazia que vai receber os tempos de execução da multiplicação das matrizes para cada n especificado

    for n in dimensao: #intera cada valor contido na lista dimensions, portanto para cada valor em n em dimensions o código vai executaro bloco de código que está dentro do laço
        # Gerar matrizes quadradas aleatórias de dimensão n x n utilizando a biblioteca numpy
        A = np.random.rand(n, n)
        B = np.random.rand(n, n)

        # Medir o tempo de multiplicação
        start_time = time.time() # a primeira parte é reponsavel por capturar o momento em que a multiplicação das matrizes esta prestes a começar que vai ser usado para calcular a diferença no tempo antes e depois da multiplicação ára encontrar o tempo gasto
        #a segunda parte retorna um número de ponto flutuante, que representa a quantidade de segundos (com frações) desde esse ponto.
        C = A @ B  # Multiplicação de matrizes, o resultado vai se armazedo em c que sera uma matriz de dimensão nxn
        end_time = time.time() #capturar o tempo após a multiplicação de matrizes ter sido concluída, depois a função retoma o tempo atual e o armazena na variavel end_time
        # Calcular o tempo de execução
        execution_time = end_time - start_time # a subtração dos dois tem como resultado o tempo ntotal gasto nas multiplicações e a primeira variavel vai armanezar o tempo total nescessário para a multiplicação de AxB
        times.append(execution_time) #adiciona o tempo de execução que foi calculado acima na lista times

        print(f"Dimensao: {n} x {n}, Tempo: {execution_time:.4f} segundos") #vai imprimir tamanho da matriz n x n tempo de execução que está formatado para 4 casas decimais
    #f-string formata a string, n insere o valor de n que é a dimensão da matriz na string e a ultima parte insere o valor de execution_time formatado em 4 casas decimais após o ponto para mostrea com precisão
    return times # retorna todos os valores de tempo a lista times que contem todos os tempos de execução para cada dimensão que foi especificada em dimensions

# Dimensões da matriz conforme especificado
dimensao = [100, 150, 200, 250]

# Executar o benchmark
times =  multiplica_matriz(dimensao)

# Plotar o gráfico log-log
plt.figure(figsize=(8, 6))  # Definir o tamanho da figura
plt.loglog(dimensao, times, marker='o', label='Log-Log')  # Gráfico log-log
plt.xlabel('Dimensão da Matriz (n)')
plt.ylabel('Tempo de Execução (s)')
plt.title('Tempo de Execução vs Dimensão da Matriz (Escala Log-Log)')
plt.grid(True)  # Adicionar grade
plt.legend()
plt.show()
```

```python
#EXERCICIO 3
#fazendo manualmente:
# Definição dos parâmetros
N = 5000  # Número de iterações
x0 = 0.1  # Valor inicial de x
a = 3.8   # valor de a para o mapeamento logistico

# Essa lista foi criada para armazenar a sequencia de valores de cada iteração de x
x_valores = [x0]

# Aqui é gerado um laço for que é responsavel por gerar a sequencia
for n in range(1, N): #ele ira executar o código n-1 vezes  gerando os valores subsequentes da sequecia
    x_n = a * x_valores[n-1] * (1 - x_valores[n-1])  # aqui é a formula do mampeamnto logisco que foi dada que é usada para calcular o proximo valor da sequencia, a paramedo de controle,  x_valores é o valor  anterior da sequencia
    x_valores.append(x_n) # essa operação mantém as sequencia dos valores gerados , cada interação de x_n é adicionado a lista  e a cada vez, o próximo valor é calculado usando o valor anterior da lista.


# Aqui é calculado a media
media = sum(x_valores) / N #tera sobre todos os elementos da lista e calcula a soma de cada um dos elementos e é dividida por N

# Calculando a variância
variancia = sum((x - media) ** 2 for x in x_valores) / (N - 1)#Somando os desvios quadrados de cada elemento em relação à média, dividindo o resultado da soma dos desvios quadrados pelo número de graus de liberdade, N-1 ajusta o cálculo da variância para amostras.


# Exibindo os resultados
print(f"Média: {media}")
print(f"Variância: {variancia}")

#USANDO NUMPY

# Definindo os parâmetros
N = 5000  # Número de iterações
x0 = 0.1  # Valor inicial de x
a = 3.8   # Escolhendo um valor de a para o mapeamento logístico

# Inicializando um array numpy para armazenar os valores de x
x_valores = np.zeros(N)
x_valores[0] = x0

# Gerando a sequência
for n in range(1, N):
    x_valores[n] = a * x_valores[n-1] * (1 - x_valores[n-1])

# Calculando a média e a variância com numpy
media = np.mean(x_valores)
variancia = np.var(x_valores)

# Exibindo os resultados
print(f"Média: {media}")
print(f"Variância: {variancia}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Tamanho da grade
N = 100

# Probabilidade de célula viva (p1) e morta (p0)
p0, p1 = 0.8, 0.2

# Criar a grade inicial aleatória usando numpy
grid = np.random.choice([0, 1], N * N, p=[p0, p1]).reshape(N, N)
#N*N representa as celulas de toda grade

# Função que calcula o próximo estado da grade
def update(frameNum, img, grid): # é responsável por calcular o próximo estado da grade a cada iteração ou seja, a cada quadro da animação
    # Criar uma cópia da grade para armazenar o próximo estado
    new_grid = grid.copy()

    # Iterar por cada célula da grade, percorre cada célula da grade e calcula quantos vizinhos vivos ela tem
    for i in range(N): #percorre as linhas i
        for j in range(N): #percorre as colunas j
            # Contar os vizinhos vivos (células adjacentes), o calculo de total de vivos é a soma das 8 células vizinhas da célula atual grid[i, j].
            total_vivos = (grid[i, (j-1)%N] + grid[i, (j+1)%N] + #horizontal
                           grid[(i-1)%N, j] + grid[(i+1)%N, j] + #vertical
                           grid[(i-1)%N, (j-1)%N] + grid[(i-1)%N, (j+1)%N] + #diagonal
                           grid[(i+1)%N, (j-1)%N] + grid[(i+1)%N, (j+1)%N]) #diagonal

            # Aplicar as regras do Jogo da Vida
            if grid[i, j] == 1:  # verificação de se a celula esta viva, se ela estiver viva fazemos a verificação do numero de vizinhos vivos para aplicar as regras
                if total_vivos < 2 or total_vivos > 3: #A célula viva morre por subpopulação. Se uma célula tem menos de dois vizinhos vivos, ela não tem recursos suficientes para sobreviver. A célula viva morre por superpopulação. Se uma célula tem mais de três vizinhos vivos, ela sofre de "congestionamento" e morre.
                    new_grid[i, j] = 0  # Morte por subpopulação ou superpopulação,na nova grade (new_grid) é atualizado para 0 (célula morta).

            else:  # Célula morta
                if total_vivos == 3: #e a célula morta tem exatamente 3 vizinhos vivos, ocorre reprodução e a célula renasce na próxima iteração.
                    new_grid[i, j] = 1  # Reprodução (célula viva)

    # Atualizar a imagem no gráfico
    img.set_data(new_grid)
    grid[:] = new_grid[:]  # Atualizar a grade original com o novo estado
    plt.title(f"Game of Life - Frame {frameNum}")
    return img,

# Configurar o gráfico
fig, ax = plt.subplots()
img = ax.imshow(grid, interpolation='nearest', cmap='pink')

# Criar a animação
ani = animation.FuncAnimation(fig, update, fargs=(img, grid), frames=200, interval=100)

# Mostrar a animação
plt.show()
```
